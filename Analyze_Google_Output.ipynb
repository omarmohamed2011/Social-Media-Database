```python
def transform_tags(tag):
    ls = []
    if 'price' in tag:
        ls.append('Price')
    if 'place' in tag:
        ls.append('Hospital')
    if 'datetime' in tag:
        ls.append('WaitTime')
    if 'Booking' in tag:
        ls.append('Reception')
    if 'insurance' in tag:
        ls.append('Reception')
    if 'structure' in tag:
        ls.append('Reception')
        ls.append('Hospital')
    if 'service' in tag:
        ls.append('Hospital')
    if 'medical_consult' in tag:
        ls.append('Doctor Feedback')
    if 'insurance' in tag:
        ls.append('Insurance Feedback')

    ls = list(set(ls))
    e = ' / '.join(str(e) for e in ls)
    return e

def transform_tags2(tag):
    ls = []
    if 'Price' in tag:
        ls.append('High Price')
    if 'Hospital' in tag:
        ls.append('Bad History with Hospital')
    if 'WaitTime' in tag:
        ls.append('Long Waiting Time')
    if 'Reception' in tag:
        ls.append('Bad History with Reception')

    ls = list(set(ls))
    e = ' / '.join(str(e) for e in ls)
    return e
```

```python
df_inq['Tag Prep'] = tags_output[451:]
```

```python
df_inq_tagged = df_inq[df_inq['Tag Prep'] != ' ']
df_inq_tagged
```

```python
units = list(set(df_inq['Business_unit']))
```

```python
units_values = []; all_dfs_ls = []
for unit in units:

    df_unit = df_inq[df_inq.Business_unit == unit]

    tags_output = []
    tags = df_unit['tags'].tolist()

    for i in range(len(tags)):
        if type(tags[i]) == str:
            out = transform_tags(tags[i])
            out = transform_tags2(out)
        else:
            out = ' '
        tags_output.append(out)
    df = df_unit[['Business_unit','Platform','Review_Text','Time']]
    df['Comment'] = tags_output

    all_dfs_ls.append(df)

    count_dict = {}
    for value in tags_output:
        if value != ' ':
            if value in count_dict:
                count_dict[value] += 1
            else:
                count_dict[value] = 1
    units_values.append(count_dict)

df_negative_reviews = pd.concat(all_dfs_ls)
df_negative_reviews = df_negative_reviews[df_negative_reviews.Comment != ' ']
```

```python
units_perc = []
for count_dict in units_values:

    total_elements = 0

    for key, value in count_dict.items():
        total_elements += value

    for key, value in count_dict.items():
        percentage = (value / total_elements) * 100
        count_dict[key] = percentage
    units_perc.append(count_dict)
```
